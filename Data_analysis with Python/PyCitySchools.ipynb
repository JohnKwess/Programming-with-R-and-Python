```python
import pandas as pd
```

```python
file1 = "Resources/schools_complete.csv"
file2 = "Resources/students_complete.csv"
```

```python
schools_data = pd.read_csv(file1)
students_data = pd.read_csv(file2)
```

```python
### District Summary
total_schools = schools_data["school_name"].count()
total_student = students_data["student_name"].count()
total_budget = schools_data["budget"].sum()
total_schools
avg_math_score = students_data["math_score"].mean()
avg_reading_score = students_data['reading_score'].mean()
math_score = students_data.loc[students_data["math_score"]>=70]["math_score"].count()
reading_score = students_data.loc[students_data["reading_score"]>=70]["reading_score"].count()
perc_math_score  = math_score / total_student
perc_reading_score = reading_score / total_student
overall_rate = students_data[(students_data["reading_score"]>=70) &(students_data["reading_score"]>=70)]["student_name"].count()/ total_student

District_data_frame = pd.DataFrame ({ "total schools" : [total_schools],
                                     "Total Students" : [total_student],
                                     "Total Budget":[total_budget],
                                     "Average Math Score" : [avg_math_score] ,
                                     "%Passing Math" : [perc_math_score],
                                    "Average Reading Score" :[avg_reading_score], 
                                     " % Passing Reading": [perc_reading_score],
                                     "Overall Passing Rate":[overall_rate]
})
                                          

District_data_frame.style.format({"Total Students" : "{:,}","Total Budget": "${:,}", "Average Math Score": "{:.2f}", "%Passing Math":"{:.1%}",
                                   "% Passing Reading": "{:.2%}",   "Overall Passing Rate": "{:.2%}"})
                           


 
                         
                          
                           
```

```python
school = schools_data.groupby("school_name").head()
```

```python
merging = pd.merge(schools_data, students_data, on="school_name")
school_info = merging.set_index("school_name").groupby(["school_name"])


types_info = schools_data.set_index("school_name")["type"].count()


total_student_schools = merging.set_index("school_name")["Student ID"].count()
Total_School_Budget = schools_data.set_index("school_name")['budget'].sum()
Per_Student_Budget = schools_data.set_index("school_name")['budget']/schools_data.set_index("school_name")["size"]
avg_math_score = school_info["math_score"].mean()
avg_reading_score = school_info["reading_score"].mean()
perc_math = merging[merging["math_score"]>=70].groupby("school_name")["Student ID"].count() /  total_student_schools
perc_reading = merging[merging["reading_score"]>=70].groupby("school_name")["Student ID"].count() /  total_student_schools
Overall_Passing_Rate  = merging[(merging["reading_score"]>=70)& (merging["math_score"]>=70)].groupby("school_name")["Student ID"].count()/total_student_schools
Overall_Passing_Rate 
school_dataframe=pd.DataFrame({"School Name":types_info,
                               "Total Students":total_student_schools ,
                               "Total School Budget": Total_School_Budget,
                              "Per Student Budget": Per_Student_Budget,
                               "Average Math Score":avg_math_score,
                               " Average Reading Score":avg_reading_score,
                               "Passing Math":perc_math ,
                               "Passing Reading":perc_reading,"Overall Passing Rate":Overall_Passing_Rate})
school_dataframe
```

```python
top_5 = school_dataframe.sort_values("Overall Passing Rate",ascending = False)
top_5.head(5)
top_5.style.format({ 'Total Students': '{:,}',
                           "Total School Budget": "${:,}", 
                           "Per Student Budget": "${:.0f}", 
                           "% Passing Math": "{:.1%}", 
                           "% Passing Reading": "{:.1%}", 
                           "Overall Passing Rate": "{:.1%}"})
                          
                           
                            


                           
                           
```

```python
### Bottom Performing Schools (By Passing Rate)
bottom_5 = school_dataframe.sort_values("Overall Passing Rate",ascending = True)
bottom_5.head()
```

```python
### Math Scores 
ninth_math = students_data.loc[students_data['grade'] == '9th'].groupby('school_name')["math_score"].mean()
tenth_math = students_data.loc[students_data['grade'] == '10th'].groupby('school_name')["math_score"].mean()
eleventh_math = students_data.loc[students_data['grade'] == '11th'].groupby('school_name')["math_score"].mean()
twelfth_math = students_data.loc[students_data['grade'] == '12th'].groupby('school_name')["math_score"].mean()

Math_scores = pd.DataFrame({
        "9th": ninth_math,
        "10th": tenth_math,
        "11th": eleventh_math,
        "12th": twelfth_math
})
Math_scores = Math_scores[['9th', '10th', '11th', '12th']]
Math_scores.index.name = "school_name"


Math_scores.style.format({'9th': '{:.1f}', 
                          "10th": '{:.1f}', 
                          "11th": "{:.1f}", 
                          "12th": "{:.1f}"})
```

```python
# reading score 
ninth_reading = students_data.loc[students_data['grade'] == '9th'].groupby('school_name')["reading_score"].mean()
tenth_reading = students_data.loc[students_data['grade'] == '10th'].groupby('school_name')["reading_score"].mean()
eleventh_reading = students_data.loc[students_data['grade'] == '11th'].groupby('school_name')["reading_score"].mean()
twelfth_reading = students_data.loc[students_data['grade'] == '12th'].groupby('school_name')["reading_score"].mean()


reading_scores = pd.DataFrame({
        "9th": ninth_reading,
        "10th": tenth_reading,
        "11th": eleventh_reading,
        "12th": twelfth_reading
})
reading_scores = reading_scores[['9th', '10th', '11th', '12th']]
reading_scores.index.name = "school_name"

#format
reading_scores.style.format({'9th': '{:.1f}', 
                             "10th": '{:.1f}', 
                             "11th": "{:.1f}", 
                             "12th": "{:.1f}"})
```
